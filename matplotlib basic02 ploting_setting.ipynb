```python
%matplotlib nbagg
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig = plt.figure()
```

```python
ax1 = fig.add_subplot(2, 2, 1)
```

```python

ax2 = fig.add_subplot(2, 2, 2) 
ax3 = fig.add_subplot(2, 2, 3)
```

```python
ax1.plot(np.random.randn(50).cumsum())
ax2.hist(np.random.randn(100), bins=20) 
ax3.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
#figure를 만들면서 그 안에 subplot을 2 * 3 형태로 만들자
fig, axes = plt.subplots(2, 3)
```

```python
axes[0,0].plot(np.random.randn(50).cumsum())
axes[0,1].hist(np.random.randn(100), bins=20) 
axes[0,2].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig = plt.figure()
plt.plot(np.random.randn(30), color="g", marker='o', linestyle="--")
```

```python
fig = plt.figure()
plt.plot(np.random.randn(30), "b.-")
```

```python
fig, axes = plt.subplots(2, 1) 
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop')) 
data.plot(kind="bar", ax=axes[0], color='k', alpha=0.7) 
data.plot(kind="barh", ax=axes[1], color='g', alpha=0.3)
```

```python
langs = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
x_pos = np.arange(len(langs))
y_pos = [10,8,6,4,2,0]

fig = plt.figure()
plt.bar(x_pos, y_pos, align='center', alpha=0.5)
plt.xticks(x_pos, langs)
plt.xlabel("Language")
plt.ylabel('Usage')
plt.title('Programming language usage')
```

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1) 
ax.plot(np.random.randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"],                     
                            rotation=30, fontsize="small")
```

```python
ax.set_title("Random walk plot") 
ax.set_xlabel("Stages") 
ax.set_ylabel("Values")
```

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label="one") 
ax.plot(np.random.randn(1000).cumsum(), "b--", label="two") 
ax.plot(np.random.randn(1000).cumsum(), "r.", label="three")
```

```python
ax.legend(loc="best")
```
